Let's modify the original bball dataframe. <br>
The row without the Player name will be removed from nba_df.<br>
The datatype is also modified in the `Number` and `Height` Column. 

```python
nba_df = bball_df.dropna(subset=['Name'])
```

Let's check that the last row is removed.

```python
nba_df.tail()
```

```python
nba_df.shape
```

This indicates that there was only one row that was missing `Name` since only one row is removed.

```python
nba_df['Number'] =nba_df['Number'].astype(int).astype(str)
```

```python
nba_df.head()
```

Now that the `Number` column represents the player number more accurately,<br>
Let's work on converting `Height` column into `cm` float value.

```python
def convert_height(df):
    for i in range(0,df.shape[0]):
        raw_string = df.loc[i,['Height']]
        feet = raw_string.str.split('-')[0][0]
        inch = raw_string.str.split('-')[0][1]                 
        nba_df.loc[i,'Height'] = int(feet)*12*2.54 + int(inch)*2.54      
    return nba_df
```

```python
nba_df =convert_height(nba_df)
```

```python
nba_df['Height'] = nba_df['Height'].astype(float)
```

### .repeat()

```python
numerical_columns = nba_df.select_dtypes('number').columns.tolist()
categorical_columns = nba_df.select_dtypes('object').columns.tolist()
```

```python
alt.Chart(nba_df).mark_bar().encode(
     alt.X(alt.repeat(), type='quantitative', bin=alt.Bin(maxbins=25)),
     y='count()').properties(width=250, height=150).repeat(numerical_columns)
```

It all shows very valuable distribution patterns.
<br>
Let's look at the shortest players.

```python
nba_df.sort_values(by='Height').head()
```

There is only one player who is under 180cm but he has made the top 25 % of the salary. <br>
This is very encouraging data for shorter players but in other words, it is extremely rare to make it to NBA if you are shorter than 180cm. 

```python
alt.Chart(nba_df).mark_point(size=10).encode(
     alt.X(alt.repeat('column'), type='quantitative'),
     alt.Y(alt.repeat('row'), type='quantitative')
 ).properties(width=120, height=120
).repeat(column=numerical_columns, row=numerical_columns)
```

Let's look at the relation of the columns. <br>
`Weight` and `Height` have postitive relation.<br>
As `Age` goes up, there are some players who make more salary and others who make less. <br>
`Height` and `Salary` do not seem to have a positive relation, but almost everybody is taller than 180cm. 

```python
alt.Chart(nba_df).mark_bar().encode(
    alt.X('count()'),
    alt.Y(alt.repeat(), type='nominal', sort='x')
).properties(width=200, height=250).repeat(categorical_columns, columns =3)
```

All teams have around 15 players (range: 13 - 19). <br>
Some numbers are more popular. <br>
Some positions have more players than others.<br>
Some colleges have produced a lot more NBA players. Here are the top 10 colleges. 

```python
nba_df.groupby('College').size().sort_values(ascending=False).head(10)
```

### . facet()

In order to look closer into `Position` which has 5 unique values, `facet()` is useful.

```python
alt.Chart(nba_df).mark_bar().encode(
    alt.X('Height', bin=alt.Bin(maxbins=30)),
    alt.Y('count()'),
    alt.Color('Position')
).properties(width=200, height=50
).facet('Position',columns=1)
```

```python
alt.Chart(nba_df).mark_bar().encode(
    alt.X('Salary', bin=alt.Bin(maxbins=30)),
    alt.Y('count()'),
    alt.Color('Position')
).properties(width=200, height=140
).facet('Position',columns=3)
```

```python
nba_df.sort_values(by='Salary', ascending=False).head()
```

The `Position` is associated with `Height` and Point Guard has the shortest players but the they all have similar pattern for the `Salary`.<br>
When you look at the top 5 highest salary players, it has three different positions. <br>
The person who made the highest salary from this NBA dataset is Kobe Bryant.
